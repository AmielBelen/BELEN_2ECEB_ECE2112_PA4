# EXPERIMENT 4 - DATA WRANGLING AND DATA VISUALIZATION

## ECE BOARD EXAM PROBLEM

```python
# Import pandas for data wrangling
import pandas as pd
```

```python
# Load the dataset
df = pd.read_excel("board2.xlsx")
df
```

```python
# Create a new column 'Average' that calculates the mean of the 4 subjects
df["Average"] = df[["Math", "Electronics", "GEAS", "Communication"]].mean(axis=1)

# Display first rows of the DataFrame to confirm the new column
print("Preview of dataset with Average column added:")
display(df.head())
```

```python
# a. Instru DataFrame: Instrumentation, Luzon, Electronics > 70
Instru = df[
    (df["Track"] == "Instrumentation") &
    (df["Hometown"] == "Luzon") &
    (df["Electronics"] > 70)
][["Name", "GEAS", "Electronics"]]

print("\n === Instru DataFrame ===")
display(Instru)
```

```python
# b. Mindy DataFrame: Female, Mindanao, Average >= 55
Mindy = df[
    (df["Hometown"] == "Mindanao") &
    (df["Gender"] == "Female") &
    (df["Average"] >= 55)
][["Name", "Track", "Electronics", "Average"]]

print("\n =========== Mindy DataFrame ===========")
display(Mindy)
```

```python
# Visualization 1: Average grade by Track
df.groupby("Track")["Average"].mean().plot(
    kind="bar", title="Average Grades by Track", ylabel="Average", xlabel="Track")
```

```python
# Visualization 2: Average grade by Gender
df.groupby("Gender")["Average"].mean().plot(
    kind="bar", title="Average Grades by Gender", ylabel="Average", xlabel="Gender", color="orange")
```

```python
# Visualization 3: Average grade by Hometown
df.groupby("Hometown")["Average"].mean().plot(
    kind="bar", title="Average Grades by Hometown", ylabel="Average", xlabel="Hometown", color="green")
```

```python
# Conclusion
print("The analysis shows that average grades vary more by Track, with smaller differences by Gender and moderate variation by Hometown, suggesting Track has the strongest influence on student performance.")
```
